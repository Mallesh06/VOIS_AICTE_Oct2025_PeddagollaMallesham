```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Airbnb_Open_Data.xlsx")
```

```python
df.isnull().sum()
df = df.dropna()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Top 10 Cities by number of listings
if "neighbourhood group" in df.columns:
    top_cities = df["neighbourhood group"].value_counts().head(10)
    plt.figure(figsize=(10,5))
    sns.barplot(x=top_cities.index, y=top_cities.values, palette="viridis")
    plt.title("Top 10 Cities by Listings")
    plt.xticks(rotation=45)
    plt.ylabel("Number of Listings")
    plt.tight_layout()
    plt.show()
```

```python
# Room Type distribution
if "room type" in df.columns:
    plt.figure(figsize=(6,6))
    df["room type"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, colors=sns.color_palette("Set2"))
    plt.title("Room Type Distribution")
    plt.ylabel("")
    plt.show()

```

```python
# Reviews Distribution
if "number of reviews" in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df["number of reviews"], bins=50, kde=False)
    plt.title("Distribution of Reviews")
    plt.xlabel("Number of Reviews")
    plt.ylabel("Count")
    plt.show()
```

```python
# Availability Analysis
if "availability 365" in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df["availability 365"], bins=50, kde=False)
    plt.title("Availability (Days per Year)")
    plt.xlabel("Days Available")
    plt.ylabel("Count")
    plt.show()
```

```python
#Save Cleaned Data
df.to_csv("airbnb_cleaned.csv", index=False)
print("\nâœ… Cleaned dataset saved as airbnb_cleaned.csv")
```
